```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def image_negative(image):
    if len(image.shape) == 3:  # RGB image
        return 255 - image
    elif len(image.shape) == 2:  # Grayscale image
        return 255 - image
    else:
        raise ValueError("Input image must be either grayscale or RGB.")

# Read grayscale image
gray_image = cv2.imread('../images/greyscale.jpeg', cv2.IMREAD_GRAYSCALE)

# Read RGB image
rgb_image = cv2.imread('../images/rgb.jpg')

# Apply image negative transformation
negative_gray = image_negative(gray_image)
negative_rgb = image_negative(rgb_image)

# Display original and negative images
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB), cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(negative_gray, cv2.COLOR_BGR2RGB), cmap='gray')
plt.title('Negative Grayscale Image')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB))
plt.title('Original RGB Image')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(negative_rgb, cv2.COLOR_BGR2RGB))
plt.title('Negative RGB Image')
plt.axis('off')

plt.show()
```
