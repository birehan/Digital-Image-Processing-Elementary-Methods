```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def gamma_correction(image, gamma=1.0):
    """
    Apply gamma correction to an input image.

    Parameters:
    - image: Input image (grayscale or RGB)
    - gamma: Gamma value for correction (default is 1.0)

    Returns:
    - Gamma corrected image
    """
    inv_gamma = 1.0 / gamma
    # Create lookup table for gamma correction
    table = np.array([((i / 255.0) ** inv_gamma) * 255 for i in np.arange(0, 256)]).astype("uint8")
    # Apply lookup table to image
    return cv2.LUT(image, table)

# Read grayscale image
gray_image = cv2.imread('../images/greyscale.jpeg', cv2.IMREAD_GRAYSCALE)

# Read RGB image
rgb_image = cv2.imread('../images/rgb.jpg')

# Apply gamma correction transformation
gamma_corrected_gray = gamma_correction(gray_image, gamma=0.5)
gamma_corrected_rgb = gamma_correction(rgb_image, gamma=2.0)

# Display original and gamma corrected images
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB), cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(gamma_corrected_gray, cv2.COLOR_BGR2RGB), cmap='gray')
plt.title('Gamma Corrected Grayscale Image (gamma=0.5)')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB))
plt.title('Original RGB Image')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(gamma_corrected_rgb, cv2.COLOR_BGR2RGB))
plt.title('Gamma Corrected RGB Image (gamma=2.0)')
plt.axis('off')

plt.show()
```
